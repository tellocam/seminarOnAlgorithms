```python
import os
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

figure_path = "../Presentation/figures/"

# Example: 4x4 grid
num_vertices = 16
adj_matrix = np.zeros((num_vertices, num_vertices))

# Connect specific vertices (adjust as needed)
connections = [(1,2), (2,7), (7,10), (10,13), (13,10), (8,4), (4,1), (13,8)]
for edge in connections:
    adj_matrix[edge[0], edge[1]] = 1
    adj_matrix[edge[1], edge[0]] = 1  # Assuming an undirected graph, set both directions

# Create a graph from the adjacency matrix
G = nx.from_numpy_matrix(adj_matrix)

# Arrange vertices in a 4x4 grid
pos = {i: (i % 4, 3 - i // 4) for i in range(num_vertices)}

# Additional information for specific vertices
vertex_info = {1: "$\mathbf{w(v_1)=18}$", 
               2: "$\mathbf{w(v_2)= 6}$",
               7: "$\mathbf{w(v_7)= 18}$",
               10: "$\mathbf{w(v_{10})= 6}$",
               13: "$\mathbf{w(v_{13})= 7}$",
               8: "  $\mathbf{w(v_{8})= 18}$",
               4: "$\mathbf{w(v_{4})= 7}$"
               }

# Plot the graph with labels
nx.draw(G, pos, with_labels=True, labels=vertex_info, font_weight='bold', node_size=700, node_color='skyblue', font_color='black', font_size=15, edge_color='gray', linewidths=1, alpha=0.7)


plt.savefig(figure_path + "ICV0.pdf")

plt.show()

```
